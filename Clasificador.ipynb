```python
import pandas as pd

# Cargar el archivo
df = pd.read_excel(r"C:\Users\Usuario\Downloads\test.xlsx")

def clasificar_norma(row):
    texto = str(row['Título']).lower() + " " + str(row['Resumen']).lower()
    
    # Palabras clave que suelen indicar derechos para víctimas
    keywords_si = ['socorro', 'asistencia', 'resarcimiento', 'protección', 'subsidio', 'reparación']
    
    if any(key in texto for key in keywords_si):
        return 1, "La norma menciona mecanismos de asistencia o reparación directa."
    else:
        return 0, "No se identifican derechos explícitos en el texto disponible."

# Aplicar la lógica
df[['caso_ok', 'Justificación']] = df.apply(
    lambda r: pd.Series(clasificar_norma(r)), axis=1
)

# Mostrar resultado
print(df)

#Guardarlo en otro archivo excel
df.to_excel(r"C:\Users\Usuario\Downloads\resultado_clasificacion.xlsx", index=False)
```

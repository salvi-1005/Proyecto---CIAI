```python
import pandas as pd
import json
import os

df = pd.read_excel(r"C:\Users\Usuario\Downloads\train.xlsx")

df = df.fillna('')

dataset = []
for index, row in df.iterrows():
    entry = {
        "messages": [
            {"role": "system", "content": "Asistente de clasificación de leyes."},
            {"role": "user", "content": f"¿De qué trata la {row['Tipo']} {row['Número']} de {row['Añosanción']}?"},
            {"role": "assistant", "content": f"Se refiere a {row['Título']}. Resumen: {row['Resumen']}. Categorizada en el tema {row['tema']}."}
        ]
    }
    dataset.append(entry)

# Guardar como JSONL
with open('entrenamiento_leyes.jsonl', 'w', encoding='utf-8') as f:
    for item in dataset:
        f.write(json.dumps(item) + '\n')
```
